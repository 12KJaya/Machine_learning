```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```
##Dataset
training_data=np.array([[0,0],[0,1],[1,0],[1,1]])
```

```
training_data
```

```
target_data=np.array([[0],[1],[1],[0]])
```

```
target_data
```

```
###Create the model
model=tf.keras.models.Sequential()
```

```
###Adding the layers
model.add(tf.keras.layers.Dense(4,input_dim=2,activation='relu'))
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
```

```
##Configuration of model
model.compile(loss='mean_squared_error',optimizer='adam',metrics=['binary_accuracy'])
```

```
##Training the model
trained_model=model.fit(training_data,target_data,epochs=2000)
```

```
model.predict(training_data).round()
```

```
#plotting loss vs epochs
loss_curve=trained_model.history['loss']
plt.title('loss vs epochs graph')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.plot(loss_curve,color='red')
```
