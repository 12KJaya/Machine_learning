```python
import cv2
```

```python
### Webcam
cam=cv2.VideoCapture(0)
while True:
    ret,frame=cam.read()
    cv2.imshow('webcam',frame)
    if cv2.waitKey(1)==13:
        break
cam.release()
cv2.destroyAllWindows()
```
